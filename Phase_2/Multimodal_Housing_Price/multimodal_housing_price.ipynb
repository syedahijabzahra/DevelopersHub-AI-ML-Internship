```python
# Multimodal ML - Housing Price Prediction Using Images + Tabular Data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Input, Concatenate, Dropout
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import warnings
warnings.filterwarnings('ignore')

np.random.seed(42)
tf.random.set_seed(42)

# ===== STEP 1: Generate Tabular Data =====
n_samples = 800

tabular_data = pd.DataFrame({
    'sqft': np.random.randint(800, 4000, n_samples),
    'bedrooms': np.random.randint(1, 6, n_samples),
    'bathrooms': np.random.randint(1, 4, n_samples),
    'age': np.random.randint(0, 50, n_samples),
    'garage': np.random.randint(0, 3, n_samples)
})

# House "quality" score drives both image brightness/color and price
quality_score = (tabular_data['sqft']/4000 * 0.4 +
                  tabular_data['bedrooms']/6 * 0.2 +
                  tabular_data['bathrooms']/4 * 0.2 +
                  (50-tabular_data['age'])/50 * 0.2)

# Price based on tabular features + some randomness
price = (tabular_data['sqft'] * 150 +
         tabular_data['bedrooms'] * 10000 +
         tabular_data['bathrooms'] * 8000 -
         tabular_data['age'] * 500 +
         tabular_data['garage'] * 5000 +
         np.random.normal(0, 15000, n_samples))
price = np.clip(price, 50000, None)

# ===== STEP 2: Generate Synthetic House "Images" =====
# Images encode quality visually (brightness/color correlates with quality_score)
def generate_house_image(quality, size=32):
    img = np.zeros((size, size, 3))
    base_color = quality  # higher quality = brighter, more uniform
    noise_level = (1 - quality) * 0.4

    img[:,:,0] = base_color * 0.6 + np.random.normal(0, noise_level, (size,size))
    img[:,:,1] = base_color * 0.7 + np.random.normal(0, noise_level, (size,size))
    img[:,:,2] = base_color * 0.8 + np.random.normal(0, noise_level, (size,size))

    img = np.clip(img, 0, 1)
    return img

images = np.array([generate_house_image(q) for q in quality_score])

print("Tabular data shape:", tabular_data.shape)
print("Images shape:", images.shape)
print("Price range: ${:.0f} - ${:.0f}".format(price.min(), price.max()))

# Show sample images
fig, axes = plt.subplots(1, 5, figsize=(15,3))
for i, ax in enumerate(axes):
    ax.imshow(images[i])
    ax.set_title(f'Price: ${price.iloc[i] if hasattr(price,"iloc") else price[i]:,.0f}')
    ax.axis('off')
plt.suptitle('Sample Synthetic House Images')
plt.tight_layout()
plt.show()

# ===== STEP 3: Preprocess =====
scaler = StandardScaler()
tabular_scaled = scaler.fit_transform(tabular_data)

# Train/test split (same indices for both modalities)
indices = np.arange(n_samples)
train_idx, test_idx = train_test_split(indices, test_size=0.2, random_state=42)

X_tab_train, X_tab_test = tabular_scaled[train_idx], tabular_scaled[test_idx]
X_img_train, X_img_test = images[train_idx], images[test_idx]
y_train, y_test = price.values[train_idx] if hasattr(price,'values') else price[train_idx], price.values[test_idx] if hasattr(price,'values') else price[test_idx]

# ===== STEP 4: Build Multimodal Model =====
# Image branch (CNN)
image_input = Input(shape=(32,32,3), name='image_input')
x = Conv2D(16, (3,3), activation='relu')(image_input)
x = MaxPooling2D((2,2))(x)
x = Conv2D(32, (3,3), activation='relu')(x)
x = MaxPooling2D((2,2))(x)
x = Flatten()(x)
image_features = Dense(32, activation='relu')(x)

# Tabular branch
tabular_input = Input(shape=(5,), name='tabular_input')
y = Dense(32, activation='relu')(tabular_input)
y = Dense(16, activation='relu')(y)
tabular_features = Dense(16, activation='relu')(y)

# Fusion
combined = Concatenate()([image_features, tabular_features])
z = Dense(32, activation='relu')(combined)
z = Dropout(0.3)(z)
output = Dense(1, activation='linear')(z)

model = Model(inputs=[image_input, tabular_input], outputs=output)
model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()

# ===== STEP 5: Train =====
history = model.fit(
    [X_img_train, X_tab_train], y_train,
    validation_split=0.1,
    epochs=30,
    batch_size=32,
    verbose=1
)

# ===== STEP 6: Evaluate =====
predictions = model.predict([X_img_test, X_tab_test]).flatten()

mae = mean_absolute_error(y_test, predictions)
rmse = np.sqrt(mean_squared_error(y_test, predictions))
r2 = r2_score(y_test, predictions)

print(f"\n=== Multimodal Model Results ===")
print(f"MAE: ${mae:,.2f}")
print(f"RMSE: ${rmse:,.2f}")
print(f"R² Score: {r2:.4f}")

# Plot training history
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Model Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.legend()

plt.subplot(1,2,2)
plt.scatter(y_test, predictions, alpha=0.5, color='steelblue')
plt.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 'r--', lw=2)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title(f'Actual vs Predicted (R²={r2:.3f})')
plt.tight_layout()
plt.show()

print("\n✅ Multimodal model (Image + Tabular fusion) complete!")
```
